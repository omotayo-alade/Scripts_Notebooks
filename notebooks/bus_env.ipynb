```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Data_clean.csv')
```

```python
df.head()
```

```python
df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
df.head()
```

```python
import matplotlib as mpl
#mpl.style.available
```

```python
mpl.style.use('bmh')

fig = plt.figure(figsize=(15,10))
ax0 = fig.add_subplot(3, 2, 1)
ax1 = fig.add_subplot(3, 2, 2)
ax2 = fig.add_subplot(3, 2, 3)
ax3 = fig.add_subplot(3, 2, 4)
ax4 = fig.add_subplot(3, 2, 5)

sns.barplot(df['Gender'].unique(), df['Gender'].value_counts(), ax = ax0)
sns.barplot(df['Age'].unique(), y = df['Age'].value_counts(), ax = ax1)
sns.barplot(df['Marital Status'].unique(), df['Marital Status'].value_counts(), ax = ax2)
sns.barplot(df['Educational Qualification'].unique(), df['Educational Qualification'].value_counts(), ax = ax3)
sns.barplot(df['Type of Firm'].unique(), df['Type of Firm'].value_counts(), ax = ax4)
plt.show()
```
