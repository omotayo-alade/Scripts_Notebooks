```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('fivethirtyeight')
%matplotlib inline
from jupyterthemes import jtplot
```

```python
df=pd.read_csv('Australia_data.csv')
df.head()
```

```python
df['Year'] = list(map(int, range(1960,2019)))
```

```python
df.set_index('Year', inplace=True)
```

```python
df.head()
```

```python
ax = sns.lineplot(data = df)
```

```python
df.reset_index(inplace = True)
```

```python
mpl.style.use('ggplot')
fig_reg = plt.figure(figsize=(15,15))
ax0 = fig_reg.add_subplot(2,2,1)
ax1 = fig_reg.add_subplot(2,2,2)
ax2 = fig_reg.add_subplot(2,2,3)
ax3 = fig_reg.add_subplot(2,2,4)

sns.regplot(x='Exchange Rate', y='GDP', data=df, color='b', ax=ax0)
sns.regplot(x='Import', y='GDP', data=df, color='g', ax=ax1)
sns.regplot(x='Export', y='GDP', data=df, color='r', ax=ax2)
sns.lineplot(x='Year', y='GDP', data=df, color='brown', ax=ax3)

ax0.set_title("Regression Plot of Exchange Rate and GDP")
ax1.set_title("Regression Plot of Import and GDP")
ax2.set_title("Regression Plot of Export and GDP")
ax3.set_title("Line Plot of GDP from 1960-2018")

plt.show()
```

```python
fig_reg = plt.figure(figsize=(15,15))
ax0 = fig_reg.add_subplot(2,2,1)
ax1 = fig_reg.add_subplot(2,2,2)
ax2 = fig_reg.add_subplot(2,2,3)
ax3 = fig_reg.add_subplot(2,2,4)

sns.residplot(x=df['Exchange Rate'], y=df['GDP'], ax=ax0)
sns.residplot(x=df['Import'], y=df['GDP'], ax=ax1)
sns.residplot(x=df['Export'], y=df['GDP'], ax=ax2)
sns.lineplot(x='Year', y='GDP', data=df, color='brown', ax=ax3)

ax0.set_title("Residual Plot of Exchange Rate and GDP")
ax1.set_title("Residual Plot of Import and GDP")
ax2.set_title("Residual Plot of Export and GDP")
ax3.set_title("Line Plot of GDP from 1960-2018")

plt.show()
```

```python
df = pd.read_csv('Data_clean.csv')
df.drop('Unnamed: 0', axis=1, inplace=True)
df.head()
```

```python
df['Age'].unique()
```

```python
age1 = df['Age'] == '16-25 Years'
age2 = df['Age'] == '26-35 Years'
age3 = df['Age'] == '36-45 Years'
age4 = df['Age'] == '46 Years above'
```

```python
df1 = pd.DataFrame(df[age1][['Efficiency', 'Innovativeness', 'Sustainability']].sum(axis=0).map(int)).transpose()
df2 = pd.DataFrame(df[age2][['Efficiency', 'Innovativeness', 'Sustainability']].sum(axis=0).map(int)).transpose()
df3 = pd.DataFrame(df[age3][['Efficiency', 'Innovativeness', 'Sustainability']].sum(axis=0).map(int)).transpose()
df4 = pd.DataFrame(df[age4][['Efficiency', 'Innovativeness', 'Sustainability']].sum(axis=0).map(int)).transpose()
```

```python
df_age = df1.append([df2, df3, df4], ignore_index=True)
```

```python
df_agegroup = df.groupby('Age')[['Efficiency', 'Innovativeness', 'Sustainability']].sum().astype(int)
df_agegroup.reset_index(inplace=True)
df_agegroup
```

```python
df_age
```

```python
df_age['Age Group'] = df['Age'].unique()
df_age.set_index('Age Group', inplace=True)
df_age.reset_index(inplace=True)
df_age
```

```python
x = df_age['Age Group']
y = [df_age['Efficiency'], df_age['Innovativeness'], df_age['Sustainability']]

plt.figure(figsize=(10,8))
plt.stackplot(x,y, labels=['Efficiency', 'Innovativeness', 'Sustainability'], baseline='zero')
plt.legend(loc='upper right')
```

```python
plt.figure(figsize=(10,8))
plt.bar(x, y[0], edgecolor='white', width=1)
plt.bar(x, y[1], bottom = y[0], edgecolor='white', width=1)
plt.bar(x, y[2], bottom = np.add(y[0],y[1]), edgecolor='white', width=1)
plt.legend(labels=['Efficiency', 'Innovativeness', 'Sustainability'], loc='upper right')
plt.show()
```
