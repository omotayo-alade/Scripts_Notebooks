```python
import pandas as pd
import numpy as np
```

```python
north = pd.read_csv('North.csv')
south = pd.read_csv('South.csv')
west = pd.read_csv('West.csv')
sales = pd.read_csv('Sales.csv')
```

```python
north.sort_values(by='Store Num', ascending=True, inplace=True)
west.sort_values(by='Store Num', ascending=True, inplace=True)
south.sort_values(by='Store Num', ascending=True, inplace=True)
sales.sort_values(by='Store Number', ascending=True, inplace=True)
```

```python
north.rename(columns={'Store Num':'Store Number', 'Lon':'Longtitude', 'Lat':'Latitude'}, inplace=True)
south.rename(columns={'Store Num':'Store Number', 'Lon':'Longtitude', 'Lat':'Latitude'}, inplace=True)
west.rename(columns={'Store Num':'Store Number', 'Lon':'Longtitude', 'Lat':'Latitude'}, inplace=True)
```

```python
north.set_index('Store Number', inplace=True)
south.set_index('Store Number', inplace=True)
west.set_index('Store Number', inplace=True)
```

```python
store = north.append([south, west], ignore_index=False)
```

```python
sales.set_index('Store Number', inplace = True)
sales.shape
```

```python
store[sales.columns] = sales
```

```python
store.reset_index(inplace=True)
store.sort_values(by='Store Number', ascending=True, inplace=True)
store.set_index('Store Number', inplace=True)
store.head()
```

```python
store['Total Sales'] = store[['Cash Sales', 'Credit Sales']].sum(axis=1)
```
